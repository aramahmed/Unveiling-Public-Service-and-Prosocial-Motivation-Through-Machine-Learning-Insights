```python
import pandas as pd

# Load the uploaded Excel file
file_path = '/content/drive/MyDrive/Perspective on the Effects of Public Service and Prosocial Motivation on Employee Engagement and Organizational Behavior/Data Collection 2024.xlsx'
xls = pd.ExcelFile(file_path)

# Display sheet names to understand the structure of the file
sheet_names = xls.sheet_names
sheet_names
```

```python
# Load the data from the first sheet
data = pd.read_excel(file_path, sheet_name='Form responses 1')

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
# Re-import necessary modules if needed
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_selection import RFE
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Assuming the data is loaded into a pandas DataFrame called 'data'
# Original Kurdish column names
original_column_names = [
    'Timestamp', 'ڕەگەز', 'تەمەن', 'ئەزموونی کارکردن', 'ئاستی خوێندەواری',
    'هیچ پۆستێکی ئیداریت هەیە', 'سەرسامم بەو کەسانەی', 'زۆر گرنگە جێ پەنجەم دیاربێ',
    'پێموایە دابینکردنی هەلی یەکسان', 'کردارو هەڵسوکەوتی جوان', 'هاوسۆزم لەگەڵ ئەو کەسانەی',
    'گرنگی دان بە خۆشگوزەرانی', 'ئامادەییم تێدایە قوربانی', 'باوەڕەم بەوە هەیە',
    'بۆ من گرنگە کە لە ڕێگەى', 'لە ڕێگەى کارەکەمەوە', 'دەمەوێت لە ڕێگەى کارەکەمەوە',
    'من دەمەوێت لە ڕێگەى کارەکەمەوە', 'ئامادەم زیاتر لەوە', 'شانازی دەکەم بە کارکردنم',
    'هەتا ئەگەر لە شوێنی تر', 'من هاوکاری سەرپەرشتیارەکەم دەکەم', 'پێشنیاری  نوێ و داهێنەرانە',
    'من خۆبەخشانە کار دەکەم', 'یارمەتی ئەو کارمەندانە', 'پشووی زیادە وەرناگرم',
    'من لە کارەکەمدا نوقم بووم', 'شانازی بەو ئیشەوە دەکەم', 'هەست بە دڵخۆشی دەکەم',
    'کارەکەم ئیلهام بەخشە بۆم', 'من بە جۆش و خرۆشم بۆ کارەکەم', 'کاتێک بەیانیان لە خەو هەڵدەستم'
]

# Corresponding English column names
english_column_names = [
    'Timestamp', 'Gender', 'Age', 'Work Experience', 'Education Level',
    'Current Administrative Position', 'Admire Civic Leaders', 'Important to Serve the Public',
    'Equality in Opportunity', 'Proper Conduct', 'Empathy for the Ill',
    'Importance of Public Welfare', 'Willingness to Sacrifice', 'Public Interest First',
    'Importance of Helping Others', 'Value of Helping Others', 'Desire to Assist Others',
    'Desire to Impact Others', 'Willingness to Overwork', 'Pride in Work',
    'Loyalty to Current Position', 'Support for Supervisor', 'Innovative Suggestions',
    'Voluntary Work', 'Help to Colleagues', 'Avoidance of Extra Work',
    'Frustration in Work', 'Pride in Results', 'Satisfaction with Work',
    'Inspirational Work', 'Passion for Work', 'Excitement to Start Work'
]

# Creating a mapping dictionary
column_mapping = dict(zip(original_column_names, english_column_names))

# Applying the new column names to the DataFrame
data.rename(columns=column_mapping, inplace=True)



# Convert Kurdish text to numeric values using Label Encoding
label_encoders = {}
for column in data.columns:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column].astype(str))


# Splitting data into features and target
features = data.iloc[:, list(range(1, 24)) + list(range(25, 29))]
target = data.iloc[:, 24]

# Encoding categorical features and target variable
le = LabelEncoder()
features = features.apply(lambda col: le.fit_transform(col.astype(str)))
target = le.fit_transform(target.astype(str))

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Feature Selection with RFE
log_reg = LogisticRegression(max_iter=1000)
selector = RFE(estimator=log_reg, n_features_to_select=10, step=1)
selector = selector.fit(X_train, y_train)
selected_features = selector.support_
X_train_selected = X_train[:, selected_features]
X_test_selected = X_test[:, selected_features]


# Assuming the same data processing and feature selection steps as previously

# Define parameter grids for Grid Search and Randomized Search
param_grids = {
    'SVM': {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto']
    },
    'KNN': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'solver': ['lbfgs', 'liblinear'],
        'penalty': ['l2']
    },
    'MLP': {
        'hidden_layer_sizes': [(50,), (100,), (100, 50)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'alpha': [0.0001, 0.001, 0.01],
        'learning_rate': ['constant', 'adaptive']
    }
}

# Instantiate models
models = {
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'MLP': MLPClassifier(max_iter=1000)
}

# Function to perform Grid Search
def perform_grid_search(model, param_grid, X_train, y_train):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_, grid_search.best_params_

# Function to perform Randomized Search
def perform_randomized_search(model, param_distributions, X_train, y_train):
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_distributions, n_iter=50, cv=5, scoring='accuracy', n_jobs=-1, random_state=42)
    random_search.fit(X_train, y_train)
    return random_search.best_estimator_, random_search.best_params_

# Dictionary to store the best models and their parameters
best_models = {}
best_params = {}
model_names = list(best_models.keys())
accuracies = []
# Running optimization for each model
for model_name, model in models.items():
    print(f"Optimizing {model_name}...")
    best_model, best_param = perform_grid_search(model, param_grids[model_name], X_train_selected, y_train)
    best_models[model_name] = best_model
    best_params[model_name] = best_param
    print(f"Best parameters for {model_name}: {best_param}")

    # Predictions and evaluation
    predictions = best_model.predict(X_test_selected)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy with Optimized Parameters: {accuracy}")
    print(classification_report(y_test, predictions))
    print("\n")
    accuracies.append(accuracy)
    # Bar chart for model accuracies
plt.figure(figsize=(10, 5))
plt.bar(model_names, accuracies, color='skyblue')
plt.title('Accuracy of Different Models After Grid Search Optimization')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()

# Line chart for model accuracies
plt.figure(figsize=(10, 5))
plt.plot(model_names, accuracies, marker='o', linestyle='-', color='blue')
plt.title('Accuracy of Different Models After Grid Search Optimization')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(True)
plt.show()

# Randomized Search can be performed similarly by calling perform_randomized_search()
```

```python
# Define colors for each model
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightpink']

# Figure 1: Bar chart with different colors for each model and labels for results
plt.figure(figsize=(10, 5))
plt.bar(model_names, accuracies, color=colors)
plt.title('Accuracy of Different Models After Grid Search Optimization')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y')

# Adding accuracy labels on top of the bars
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 0.02, f'{accuracy:.3f}', ha='center')

plt.savefig("figure_1_labeled.png")
plt.show()

# Figure 2: Line chart with different colors for each point and labels for results
plt.figure(figsize=(10, 5))
plt.plot(model_names, accuracies, marker='o', linestyle='-', color='blue')
plt.title('Accuracy of Different Models After Grid Search Optimization')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(True)

# Adding accuracy labels on the line chart
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 0.02, f'{accuracy:.3f}', ha='center')

plt.savefig("figure_2_labeled.png")
plt.show()
```
